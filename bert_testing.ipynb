## 使用bert來觀察句向量


以下模型使用方式可以參考此網址：https://github.com/hanxiao/bert-as-service  
這裡使用bert serving client作為線上伺服器來使用bert模型  
模型經過fine-tune (使用整理過的資料集)

### 載入套件

```python
# 載入Bert線上伺服器
from bert_serving.client import BertClient
# 簡體繁體轉換工具
from opencc import OpenCC

# PCA套件
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 使用bert-as-service線上伺服器服務
bc = BertClient()
```

### 輸入想觀察的句子

```python
sentence = ['一個假裝快樂',
            '一個真心難過']
```

```python
result = bc.encode(sentence)
```

### 句向量

```python
result[0]
```

```python
result[1]
```

### 透過cosine similarity來簡單判斷兩句話是否在向量空間中有一定的關係

```python
from sklearn.metrics.pairwise import cosine_similarity
```

### 若在丟入cosine similarity前沒有進行reshpae會報錯

```python
cosine_result = cosine_similarity(result[0],result[1])
```

```python
cosine_result = cosine_similarity(result[0].reshape(1,-1),result[1].reshape(1,-1))
```

```python
print(cosine_result)
```

### 小結

句子一：一個假裝快樂  
句子二：一個真心難過  

cosine similarity：0.8964  
代表這兩句話在向量空間中的相似程度有0.896
